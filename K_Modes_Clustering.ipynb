## K-Modes Model

- K-Modes is an algorithm for Clustering Categorical Data.

  - It is used to partition a dataset into a specified number of Clusters, where each Cluster is characterized by a Mode ⟶ The Most Frequent Categorical value in the Cluster

  - The Distances between data objects are calculated using the Haming Distance, a Dissimilarity Measure ⟶ Numerical Measure of how different two objects are

```python
# install kmodes
!pip install kmodes
```

```python
# import kmodes
from kmodes.kmodes import KModes
```

```python
# list to store model costs
costs = []

# list of K values
Ks = [2, 3, 4, 5, 6]

# for each value of k
for k in Ks:

  # build KModes model with k clusters
  kmode_model = KModes(n_clusters=k, init='random', n_init=5, verbose=1)
  # predict clusters
  kmode_model.fit_predict(features)
  # append model cost
  costs.append(kmode_model.cost_)


# print costs
print('\n', f'Costs (Summ of All Dissimilarities between Clusters): {costs}')
```

```python
# plot elbow
fig, axs = plt.subplots(figsize=(8, 6))

plt.plot(Ks, costs, color='green')

plt.xlabel('# Clusters')
plt.ylabel('Cost')

plt.title('Best K: Elbow Method', pad=13, size=15)

plt.show()
```

- The Point of Inflexion of the Curve, shows the Optimal Number of Clusters for the Personal Growth Survey Dataset

  - $K = 5$

```python
# build K-Mode model with k=5
kmode_final = KModes(n_clusters=5, init='random', n_init=5, verbose=1)

# predicted clusters
clusters_f = kmode_final.fit_predict(features)
print('\n', 'Predicted Clusters:')
print(clusters_f)
```

```python
# cluster centroids
centroids_f = kmode_final.cluster_centroids_

print('Cluster Centroids:', '\n')
for el in centroids_f:
  print(el, '\n')
```

```python
# add predicted clusters to df
pw_df['K-Modes_clusters'] = clusters_f

pw_df
```

### Multiple Correspondence Analysis (MCA)

```python
!pip install --user prince
import prince
```

```python
# create instance of MCA
mca_pw = prince.MCA(n_components=2,
                    n_iter=5,
                    copy=True,
                    check_input=True,
                    engine='sklearn'
                    )

mca_pw
```

```python
# fit data
mca_pw = mca_pw.fit(features)

# transform data
dim_redu = mca_pw.transform(features)

dim_redu
```

```python
# plot
fig, axs = plt.subplots(figsize=(8, 6))

sns.scatterplot(x=dim_redu[0], y=dim_redu[1], hue=clusters_f, palette='Paired')

plt.xlabel('$x_1$')
plt.ylabel('$x_2$')

plt.title('Personal Growth Survey - K-Mode Clustering', pad=13, size=15)

plt.show()
```
